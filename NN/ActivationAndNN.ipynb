```
import numpy as np
import matplotlib.pyplot as plt
```

```
x=np.linspace(-10,10,1000)
```

```
y=np.where(x>0,x,0)
```

```
plt.plot(x,y)
plt.title("Activation Relu")
plt.grid(True)
```

```
y_Sig=1/(1+np.exp(-x))
```

```
plt.plot(x,y_Sig)
plt.title("Activation Sigmoid")
plt.grid(True)
```

```
y_tanh=(np.exp(x)-np.exp(-x))/(np.exp(x)+np.exp(-x))
```

```
plt.plot(x,y_tanh)
plt.title("Activation tanh")
plt.grid(True)
```

**09/02/2021 Excercise **
Using the data below (Xnew features and ynew binary outputs as groundtruth) create a sequencial neural netowrk model with three layers:
Layer 1: 100 neurons
Layer 2: 10 neurons
Layer3 (output layer): 1 neuron
output layer activation : sigmoid
loss: binary_crossentropy
optimizer: adam
metrics: accuracy
epochs: 100

Come out with an accuracy score for this network.

```
Xnew=np.random.random((1000,10))
ynew=np.random.random((1000,1))
ynew=np.where(y>0.5,1,0)
```

```
model = Sequential()
model.add(Dense(100,input_dim=10))
model.add(Dense(10))
model.add(Dense(1))
model.add(Activation('sigmoid'))
```

```
model.compile(loss='binary_crossentropy',optimizer='adam', metrics=['accuracy'])
```

```
model.fit(Xnew,ynew,epochs=100)
```

```
class_probabilities = model.predict(Xnew)
print (class_probabilities)
```

```
classes = np.where(class_probabilities>0.5,1,0)
print (classes)
```

```
print ("Accuracy: ",accuracy_score(ynew,classes))
```
